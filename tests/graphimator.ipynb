```python
initial_state = {'a': -0.45, 'b': -0.05, 'c': 0}
initial_point = (-0.4, initial_state)
# points are now of the form (t, state)

def da_dt(t, state):
        return state['a'] + 1
def db_dt(t, state):
        return state['a'] + state['b']
def dc_dt(t, state):
        return 2 * state['b'] + 3 * t
derivatives = {
    'a': da_dt,
    'b': db_dt,
    'c': dc_dt
}
```
