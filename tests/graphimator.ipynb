**Imports**

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.insert(1, sys.path[0].replace('tests', 'src'))
from gen_estimator import *
```

**initialization**

```python
def da_dt(t, state):
    return state['a'] + 1


def db_dt(t, state):
    return state['a'] + state['b']


def dc_dt(t, state):
    return 2 * state['b'] + 3 * t


derivatives = {
    'a': da_dt,
    'b': db_dt,
    'c': dc_dt
}

initial_state = {'a': -0.45, 'b': -0.05, 'c': 0}
```

**graphing**

```python
l = gen_estimator(derivatives)
for i in range(1,6):
    print('i is ',i)
    m = l.calc_estimated_points((i,initial_state),step_size=0.001,num_steps = 10000)
    keys = m[0][1].keys()
    dict = {}
    state = []
    for key in keys:
        dict[key] = []
        for i in range(len(m)):
            dict[key].append(m[i][1][key])
    for i in range(len(m)):
        state.append(m[i][0])
    for j in keys:
        plt.plot(state,dict[j],label = j)
    plt.xlim(0,10)
    plt.ylim(0,10)
    plt.legend()
    plt.show()
```
